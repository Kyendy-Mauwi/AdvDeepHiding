# Questions 
what location has the fastest rollercoaster(min of 10)

```python
df['Location'].value_counts()
```

```python
#others isn't a location
df.query('Location != "Other"')\
.groupby('Location')['Speed_MPH']\
.agg(['mean', 'count'])\
.query('count >=10')\
.sort_values('mean')
```

```python
ax = df.query('Location != "Other"')\
.groupby('Location')['Speed_MPH']\
.agg(['mean', 'count'])\
.query('count >=10')\
.sort_values('mean')['mean']\
.plot(kind='barh', title='Avg Coaster speed by location', figsize=(12, 5))
ax.set_xlabel('Avg Coaster speed')
plt.show()
```
